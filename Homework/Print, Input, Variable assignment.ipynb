If I want to print "eggplant is my favourite food" to the screen, I can do it like this:

````python
print("eggplant is my favourite food")
````

Variables are useful for saving information in our program. 

To create a variable and assign a value to it, we do the following:

````python
favourite_food = "eggplant"
````

Note that the variable name is on the left side of the = sign, and the variable value is on the right side of the = sign. 

Don't forget the naming conventions we learned about in class!

The input() function is useful for asking the user questions. 

For example, if you want to ask the user what their favourite food is, you would write:

````python
input("What is your favourite food?")
````

We can also use the print() function, variable assignment, and the input() function together!

For example, if I want to greet the program user, I could do:

````python
user_name = input("Hi! What's your name? ")
print("Hi", user_name)
````

In the code above, I obtained an input from the program user, assigned it to the variable `user_name`, and printed it with a greeting.
