# While Loops

When you are asking for your friend's attention, you might have to call them several times until they realize you're calling them. You may choose to continue repeating their name, until they respond to you. 

In other terms, **while** your friend has not responded, you continue calling their name. 


Screen Shot 2021-11-16 at 4.41.46 PM.png

This describes a while loop, which will continue to repeat until a desired condition is achieved. In the case of the example above, the condition is achieved when your friend notices that you called their name!

The format of a while loop is as follows:
```python
while desired_condition_not_achieved:
  # perform some operation
```

For our example above, the while loop statement might look something like this:

```python
while friend_not_responding:
  print("friend's name") #or yell their name!
```

We don't know how many times we'll need to call their name before they respond, which is what differentiates this from a for loop. In a for loop, we already know ahead of time how many times we need to repeat the code. Here, who knows when our friend will hear us!

Notice how the conditional statement is very similar to the type of conditional statements written for if statements!
