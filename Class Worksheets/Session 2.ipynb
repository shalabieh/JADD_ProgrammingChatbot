##What's a String?

A string is a variable type in python, that is useful when we want to work with words and sentences. For example, if you want to initialize your name as a variable, you can do it as a string:

```python
name = 'Sarah'
```
You can also do it this way:
```python
name = "Sarah"
```

Notice the use of the single quotation marks (' ') vs. the double quotation marks (" "). It doesn't matter which ones you use, as long as you **use the same type of quotation marks for defining your string on both ends**.

A string is made up of characters, that are the individual letters. For example, the name Sarah is made up of the characters 'S', 'a', 'r', 'a', 'h'.

You can imagine that each letter in a string is a building block, similar to Lego! Each Lego brick ðŸ§± is important for the final structure. 


A string doesn't have to be just one word, it can also be a sentence:
```python
greeting = 'hello world!'
```

In this string, the space that separates the word 'hello' from the word 'world!' is also a character.

##Slicing strings

Sometimes we want to access part of a string, but not the entire string. We can **slice** into a string to access certain characters in the string. 

First, we want to determine how long strings are. Python counts each character in a string as one unit of length. To calculate the length of the string, we use the **len()** function:


````python
s1 = 'happy'
string_length = len(s1)
print(string_length)
````

This indicates that there are 5 characters that make up the string `'happy'`

What happens if we want to obtain more than one character from our string? We can slice the string into a chunk, just like a cake!

We need to define a range of indices that we want to slice the string. In python, to represent a range, we use the ":" character. Let's try using this character to slice a string. 

````python
#Guess what the ouput would be
s1 = 'happy'
print(s1[0:3])
````

When we specify a range for slicing, it includes the first number, but does not include the outermost part of the range. For example, a slice of [1:4] will produce the following indices: 1, 2, and 3, but **not** 4!

#String Methods

Python provides several useful methods for manipulating strings, so that we don't have to write functions that do them!

There are many methods, and we're going to familiarize ourselves with some of them. We can display all of the string methods by calling the help function:

````python
help(str)
````

A method will take an already-initalized string, and return a **new value** once the operation is done. **The method does not modify the existing string**.

To call a string method, you type the name of the string, followed by a '.', and the name of the method, and round brackets (). If the method requires an input, you type that input into the round brackets. For example:
```python
s1 = 'happy'
s2 = s1.capitalize()
```

### Casefold

This method returns a string where all characters are lower case.

````python
welcome = "Welcome To Python Programming a CHATBOT!"

welcome_method = welcome.casefold()

print("string before applying casefold method: " + welcome)
print("string after applying casefold method: " + welcome_method)
````

note several things in the example above:
1. the 'welcome' string did not change after calling the method
2. at the end of the method, we include round brackets (), 

### Endswith

This method returns **True** if the string ends with the specified value (put in quotation marks), otherwise it returns **False**.

Run the following examples to find out more about it!

````python
statement = "would you like to play with me?"
question = statement.endswith("?")
print(question)
````

````python
statement = "would you like to play with me?"
question = statement.endswith("!")
print(question)
````

### Rfind  

When you're getting to know somebody, one of the first things you might do is ask for the person's name. Having obtained their name, you can respond with a personalized greeting, like the one displayed above. 

To extract their name, we need to predict the different ways they might introduce themselves. If your program asks someone what their name is, they will likely respond like this:

1. "My name is Daniel", or
2. "I'm Daniel", or
3. "Daniel"

In cases 1 and 2, how could we isolate the name 'Daniel' from the rest of the response? 

The name is always the last word in the response, so if we can find the index of the last space, we can slice everything after that! using the rfind() method, we can find the rightmost index of the space character:

````python
response = 'my name is Daniel'
space_index = response.rfind(' ')
#slice everything after that space
name = response[?:?]
print("nice to meet you " + name + "!")
````

So, using the rfind() method, and the ability to slice strings, we are able to print a personalized greeting to someone!
