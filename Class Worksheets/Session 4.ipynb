#Functions

There are two types of function: 
1. Built-in Functions and 
2. Self-Defined Functions. 

Python has built-in functions in it that are always avaliable, like **print()** and **input()**. You can directly call them when you want to use it. 

We are going to talk about how to define our own functions and use them in this session.

## How to create a function?
Functions are created using the keyword "**def**".

The following code is an example of function definition. It might be a little bit confusing now, but will get clear after our explanation!

````python
def double_calc(original: int) -> int:
  """
  Return the doubled value of input value 'original'.
  >>>double_calc(8)
  16
  >>>double_calc(4)
  8
  >>>double_calc(5.2)
  10.4
  """
  answer = 2*original
  return answer
````

When we ran the above function, why was there no output generated? When you run a function in Python, it stores the information of that function, and waits for you to **call** the function with an input. In the code above, we did not provide any numerical input into the function, so it could not calculate anything! We need to call the function and provide an input for the function to give us an output. It can be done like this:

````python
double_calc(9)
````

In the example above, we did not save the function output to a variable. However, the `double_calc` function outputs an integer, which can be stored in a variable and used later on.

### Headline
The headline of a function is using the following format:

```python
def function_name (input_variable_name: input_variable_type) -> output_variable_type:
```
Here our function name is "double_calc", input variable name is "calc", input variable type is "int" (integer), and output variable type is also integer. So we have the headline as shown in above code:
```python
def double_calc(original: int) -> int:
```

Note: the colon after the headline is necessary!

### DocStrings
It's useful to explain what exactly our function is trying to do, especially if somebody else is reading through and using our code. The DocStrings provide us the opportunity to explain what our function can do, and show how to use the function with expected outputs. 

To define a DocStrings, simply include 3 apostrophes at the beginning `'''` and 3 apostrophes at the end `'''`.

To show how the function is called, we start the line with >>>. The line that follows that immediately is the function output.

```python
  '''
  Return the doubled value of input value 'original'.
  >>>double_calc(8)
  16
  >>>double_calc(4)
  8
  >>>double_calc(5.2)
  10.4
  '''
```

### Function Body

The body is where we actually write the code that will take our inputs, and turn them into the output. 

### Return

In the examples above, we see something called a `return`. This indicates what is outputted at the end of the function. The headline usually indicates the variable type of what is being outputted. So, what follows the return should be of that variable type. For example, in the `full_name` function, we are expecting a string output. The return should then be a string. 

Here is another example for self-defined functions. Try to explan it by yourself!

Note: We run it after the definition to test the function!

````python
def full_name(first_name: str, last_name: str) -> str:
  """
  Return the full name according to the first name and last name.
  >>>full_name('Jack', 'Smith')
  Jack Smith
  >>>full_name('Carmen', 'Chen')
  Carmen Chen
  """
  whole_name = first_name + ' ' + last_name
  return whole_name
````

````python
full_name('Carmen','Han')
````

## Function Design Recipe
These examples can be summarized by a function design recipe:
>1. Header: def + function name + parameters + type contract.
>
>2. Examples: Name the function and figure out its role, input and output.
>
>3. Description: Natural Language describing what the function does. You need to: mention each parameter by name, describe the return value, explain the precondition.
>
>4. Body: Concrete implementation code.
>
>5. Test: Call your functions and compare the return values to the expected value.

## Variables Within Functions

We sometimes define variables that are used within our function. Let's see what happens when we try to print the value of a variable outside of the function it is in:


````python
def full_name(first_name: str, last_name: str) -> str:
  """
  Return the full name in correct format according to the first name and last name.
  >>>full_name('Jack', 'Smith')
  Jack Smith
  >>>full_name('Carmen', 'Chen')
  Carmen Chen
  """
  whole_name = first_name + ' ' + last_name
  return whole_name
````

````python
full_name('Carmen','Han')
print(whole_name)
````
