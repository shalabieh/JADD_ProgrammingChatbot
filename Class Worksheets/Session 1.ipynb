#Process: Variables 

When we get information from someone, we like to remember that information so that we can refer to it later. Imagine what would happen if we couldn't remember anything anyone ever told us! 

We need to teach our chatbot to also remember details, by storing that information! 

A variable is a symbol for storing information. It is a container to hold information. Instead of just printing to the screen, variables we define will be kept in the memory and can be referenced and manipulated in a computer program.

##Creating Variables

Variables label data with a descriptive name, which makes our program more understandable. To create a variable, we have 2 steps:
1. Make a suitable name for it
2. Assign a value to it. 

The equal sign (=) is used for assigning values to variables. The format is: 
        
              variable name =  variable value


In this session we are going to assign words or sentences as variables values, which are called *string variables*. 

````python
# for example:
day = "Thursday"
weather = "It is raining outside today"
````

###Rules for variable name
1.
A variable name must start with a letter or the underscore character (_). You cannot start with numbers.


2.
A variable name can only contain alpha characters, digits and underscores (A-z, 0-9, and _ ). We can not include any space in a variable name!

3.
Variable names are case-sensitive (age, Age and AGE are three different variables)

>Good examples of variable names: age, pet_name, favourite_game...

##Printing Variables
We can print the word or sentence stored inside of a variable, using the print() function we learned before:

````python
my_favourite_colour = 'blue'
print(my_favourite_colour)
````

Because `my_favourite_colour` is not in quotation marks, Python looks for a variable named `my_favourite_colour` rather than literally printing "my_favourite_colour".

What happens if we put `my_favourite_colour` in quotations?

````python
print('my_favourite_colour')
````

What if we want to print something more meaningful, such as "my favourite colour is blue"? We can do it this way:

````python
my_favourite_colour = 'blue'
print("my favourite colour is ", my_favourite_colour)
````
