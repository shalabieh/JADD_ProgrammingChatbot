# If Statement 

We can make decisions is by comparing two things. For example, is something cheaper, larger, or heavier? We can also compare to something we already know.  

We can teach our computers to tell the relationship between two numbers: is A larger than B? Is A smallar than B? or is A equal to B...

This is a conversation that you could potentially have with your friend! How would you respond if they said they're feeling happy? 

Can we get our chatbot to display a similar behaviour? Let's try!

````python
feeling = input("Hey, How are you feeling today?")

if feeling == "happy":
  print("that's great to hear!")
````

Let's breakdown exactly what this code is trying to do:

if statement breakdown.png

So, an if statement allows us to *process* the information we are receiving from the program user!

Hmmm... in the code above, when we want to process information, we used a double equal (`==`) rather than the single one we use to assign variables!

Why is that? 

Well, in Python, if we want to assign a value to a variable, we use a single `=`. In an if statement, we are checking to see if the value on the left is *equal to* the value on the right, so we use the `==`!

````python
# we can compare two strings to each other
print('python' == 'python')
````

````python
# We can compare the values stored in two variables to each other
name_one = 'Evan'
name_two = 'Cassandra'

print(name_one == name_two)
````

````python
# We can compare the value stored in a variable to a string
game = 'Minecraft'

print(game == 'Minecraft')
````

In the example above, we introduced a new term: 
```python
if
```
This term helps us determine **if** we meet a condition, and is often called a **conditional statement**. 



Notice that we applied the following formula:

# If, Elif, Else

How can we translate what we are feeling into satisfying actions? We can use a logical process.

# OR operator in if statements

What if we have multiple acceptable options for a condition? For example, "happy" is not the only positive feeling word! Other words may be "excited", "great", etc. How can we make sure our chatbot can handle different, but similar, responses from the user? 

Let's think about it:
1.  If the user says they're happy **or** they're excited **or** they're great  chatbot responds "that's great!"
2.  If the user says they're sad **or** they're made **or** they're bad  chatbot responds "I'm sorry to hear that"

Notice the word **or** used above. It indicates that any of these responses meet the condition.

Let's see how we can integrate that into our code:

````python
feeling = input("Hey, How are you feeling today?")

if feeling == "happy":
  print("that's great to hear!")
elif feeling == "sad":
  print("I'm sorry to hear that.")
else:
  print("Sorry I don't understand what you mean.")
````

Note that the user inputs either *happy* or *great*, for the chatbot to respond "That's great!".
